```python
def find_outliers_iqr(data, column, factor=1):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

outliers_ibm = find_outliers_iqr(ibm, 'DailyRate')
outliers_ibm
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test  = sc.fit_transform(x_test)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
x_train_res, y_train_res = sm.fit_resample(x_train, y_train.ravel())
```

```python
print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))

print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
selected_features = ['Age','MonthlyIncome','TotalWorkingYears', 'YearsInCurrentRole', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
x = ibm[selected_features].values
```

```python
from sklearn.cluster import KMeans
#elbow method graph
wcss = []
for i in range(1,11):
    kmeans=KMeans(n_clusters= i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show
```

```python
kmeans = KMeans(n_clusters= 3, init='k-means++', max_iter=300, n_init=10, random_state=0)
```

```python
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

```python
ibm['Cluster'] = y_kmeans
```

```python
ibm.groupby('Cluster').mean()
```

```python
#printing centroids
kmeans = KMeans(n_clusters=3).fit(x)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s=50, c='purple', label = 'Experienced employees')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s=50, c='orange', label = 'Mid-Level employees')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s=50, c='red', label = 'Inexperienced employees')

plt.scatter(kmeans.cluster_centers_[:, 0] , kmeans.cluster_centers_[:, 1] , s=50, c='yellow', label = 'Centroids')
plt.title('Features')
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(x)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples = 7, metric='euclidean')
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

neigh = NearestNeighbors(n_neighbors=7)
nbrs = neigh.fit(X)  # Make sure X is scaled!
distances, indices = nbrs.kneighbors(X)

distances = np.sort(distances[:, 6])  # 4 because min_samples=5
plt.plot(distances)
plt.ylabel("Nearest Neighbor Distance")
plt.xlabel("Points sorted by distance")
plt.title("k-distance graph for DBSCAN")
plt.grid(True)
plt.show()
```

```python
clusters = dbscan.fit_predict(X)
```

```python
clusters = dbscan.labels_
```

```python
from pandas import Series, DataFrame
ibm['cluster group'] = Series(clusters, index=ibm.index)
ibm
```
